```python
# calculate a face embedding for each face in the dataset using facenet
from numpy import load
from numpy import expand_dims
from numpy import asarray
from numpy import savez_compressed
from keras.models import load_model
import cv2
import matplotlib.pyplot as plt
import numpy as np
import timeit
from timeit import default_timer as timer



# get the face embedding for one face
#dapetin fitur pixel wajah
temp = 0
def get_embedding(model, face_pixels):
	# scale pixel values
	face_pixels = face_pixels.astype('float32')
	# standardize pixel values across channels (global)
	mean, std = face_pixels.mean(), face_pixels.std()
	face_pixels = (face_pixels - mean) / std
	# transform face into one sample
	samples = expand_dims(face_pixels, axis=0)
	# make prediction to get embedding
	yhat = model.predict(samples)
	return yhat[0]

# load the face dataset
data = load('FaceNet+RetinaFace+Faces96+Lengkap100.npz') # model yang digunakan untuk detect wajah
#output nya di tampung ulang train dan test nya
#trainX, trainy, testX, testy = data['arr_0'], data['arr_1'], data['arr_2'], data['arr_3']
timeX,timey = data['arr_6'], data['arr_7']
print('Loaded: ',timeX.shape, timey.shape)

model = load_model('facenet_keras.h5') # arcface dengan input 112 ,112 , 3


print('Loaded Model')

start3 = timeit.default_timer()
newTimeX= list()
for face_pixels in timeX:
	embedding= get_embedding(model,face_pixels)
	newTimeX.append(embedding)

newTimeX= asarray(newTimeX)


stop = timeit.default_timer()
print(newTimeX.shape)

jumlah = stop-start3
print('elapsed time in seconds in feature extraction embedding wajah  FaceNet + RetinaFace dengan 100 gambar : %f' % (jumlah))
```
